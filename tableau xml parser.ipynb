```python
# Import xml parser
from xml.etree import cElementTree as ET

# Enter the file path of the tableau workbook below:
tableau_workbook = "C:\\Users\\chbrooks\\Desktop\\audit\\new product\\stat sum replacement - DB2.twb"

technical_doc_name = tableau_workbook[:-4]+" technical document.txt"

with open(technical_doc_name,"w") as tech_doc:

    # Feed tableau workbook to xml parser
    tree = ET.parse(tableau_workbook)
    root = tree.getroot()

    # Prints out information about database connection
    tech_doc.write("---------------------------------------------------------\n")
    tech_doc.write("Database Connection:\n")
    tech_doc.write("---------------------------------------------------------\n\n")
    print_list = ("username", "dbname", "server", "class", "port", "schema")
    for item in print_list:
        printable_item = item+":"+root[3][1][0][0][0][0].attrib[item]
        tech_doc.write(printable_item)
        tech_doc.write("\n")
    tech_doc.write("\n\n")

    # Prints out the SQL statement used to query the database
    tech_doc.write("---------------------------------------------------------\n")
    tech_doc.write("SQL:\n")
    tech_doc.write("---------------------------------------------------------\n\n")
    tech_doc.write(root[3][1][0][1].text)
    tech_doc.write("\n\n\n")

    # Loops through a dictionary of calculation names and replaces them with their 
    # assigned names in the text
    def calc_renamer(text, dic):
      for a, b in dic.iteritems():
        text = text.replace(a, b)
      return text

    # Gets a list of the calculated fields and creates a dictionary to be used to 
    # replace calculated field numbers with their given names
    calc_dict = {}
    for child in root[3][1].iter('column'):
        for subchild in child.iter("calculation"):
            if "Calculation" in child.attrib["name"]:
                caption = "["+child.attrib["caption"]+"]"
                calc_dict.update({child.attrib["name"]:caption})




    # Print out all of the calculated fields and the code used to create them
    tech_doc.write("---------------------------------------------------------\n")
    tech_doc.write("Calculated fields: \n")
    tech_doc.write("---------------------------------------------------------\n\n")
    for child in root[3][1].iter('column'):
        for subchild in child.iter("calculation"):

            if "Calculation" in child.attrib["name"]:    
                tech_doc.write(child.attrib["caption"])
                tech_doc.write("\n\n")
                tech_doc.write(calc_renamer(subchild.attrib["formula"],calc_dict))
                tech_doc.write("\n\n")
                tech_doc.write("\n\n")
                tech_doc.write("---------------------------------------------------------")
                tech_doc.write("\n\n")

tech_doc.close()

with open(technical_doc_name, "r") as tech_doc:
    print tech_doc.read()
```
